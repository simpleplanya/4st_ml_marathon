- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np 
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
# 測試不同群數的影響
estimators = [('k_means_3', KMeans(n_clusters=3)),
              ('k_means_2', KMeans(n_clusters=2)),
              ('k_means_1', KMeans(n_clusters=1,))]
fignum = 1
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4, 3))
    ## fit data
    est.fit(X)
    
    labels = est.labels_
    plt.title(name)
    plt.scatter(X[:, 0], X[:, 1],
               c=labels.astype(np.float), edgecolor='k')
    plt.show()
```

```python
# 觀察n_init參數影響
estimators = [('n_init_1', KMeans(n_init=1)),
              ('n_init_5', KMeans(n_init=5)),
              ('n_init_100', KMeans(n_init=100))]
fignum = 1
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4, 3))
    ## fit data
    est.fit(X)
    
    labels = est.labels_
    plt.title(name)
    plt.scatter(X[:, 0], X[:, 1],
               c=labels.astype(np.float), edgecolor='k')
    plt.show()
```
